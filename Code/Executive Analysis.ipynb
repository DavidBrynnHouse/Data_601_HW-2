# Analysis of how Humidity and Temperature are Related
This note book contains an analysis of 1343 different U.S cities and their recorded temperature and humidity at about 7pm on October 5, 2020. Data was obtained from the Open Weather API. A correlation of -.665 was determined for our collected data.

---

## Temperature and Humidity
Tempurature is a measure of the average energy of a medium, in this case the outdoor air. Humidity is the measure of how much water is in the air. Both of these values are important for their description of the weather on a certain day. 

---

## Goal
My goal was to determine if there is a correlation between temperature and humidity. If there is a strong enough correlation between the two values then if you have access to one value but not the other you would be able to infer a value.

---

## Methods
* Obtained a list of U.S cities from [this site](https://www.britannica.com/topic/list-of-cities-and-towns-in-the-United-States-2023068).
* Saved that list as a CSV File.
* Using the Open weather API retrieved the current temperature and humidity for each city.
* Plotted a scatterplot of the data and calculated the correlation.

```python
import pandas as pd

temp_humidity_df = pd.read_csv("../Data/temp_humidity.csv")
temp_humidity_df.plot(x='Temperature', y='Humidity', kind='scatter')
temp_humidity_df.corr()
```

## Conclusion
As we can see from the scatter plot there is a negative correlation between the temperature and humidity in a city. This value was calculated at -.665 which means there is a fairly strong correlation between the two values. We can conclude then that there is a relationship between the two variables. We cannot conclude if it is temperature impacting humidity or the other way around since this is a correlation analysis. The results of this analysis are interesting because [previous research](https://trustyjoe.com/how-does-humidity-affect-temperature-in-your-house/) has shown that warm air holds more moisture. Our results may be due to the time of year, since the changing seasons could have a large impact on temperature and humidity. Further research might perform similar analysis throughout the year to determine if this result changes.
